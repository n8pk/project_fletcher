```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import re

from sklearn.manifold import TSNE
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import Normalizer
from sklearn.decomposition import TruncatedSVD
from sklearn.decomposition import NMF

import plotly.plotly as py
import plotly.graph_objs as go

%pylab inline
```

```python
preprocessed = pd.read_pickle('preprocessed.pkl')

links = pd.read_pickle('pkl/mvp_links.pkl')
```

# strain name filter

```python
# after some EDA, the names of strains are showing up way too frequently
# we need to parse these out

names = []
for name in links:
    name = name.split('/')[2]
    name = name.split('-')
    for n in name:
        names.append(n)
names[:10]
names += ['girl', 'scout', 'cookie', 'gorilla', 'glue', 'strain']
```

```python
name_filtered = []
for strain in preprocessed:
    check = ''
    for word in strain.split():
        if word not in names:
            check += word + ' '
    name_filtered.append(check)
```

# NMF

```python
df = pd.DataFrame(data = name_filtered, index = [link.split('/')[2] for link in links])
df.head()
```

```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df = 0.6, min_df = 3, stop_words = 'english', ngram_range = (2,2))
tfidf = tfidf_vectorizer.fit_transform(name_filtered)
```

```python
tfidf_vectorizer.get_feature_names()[:5]
```

```python
nmf = NMF(n_components = 4, random_state = 42, alpha = 0, l1_ratio = 0).fit(tfidf)
```

```python
tfidf_feature_names = tfidf_vectorizer.get_feature_names()
display_topics(nmf, tfidf_feature_names, 10)
```

```python
nmf_embedding = nmf.transform(tfidf)
nmf_embedding = (nmf_embedding - nmf_embedding.mean(axis=0))/nmf_embedding.std(axis=0)
```

```python
top_idx = np.argsort(nmf_embedding,axis=0)[-6:]

count = 0
for idxs in top_idx.T: 
    print("\nTopic {}:".format(count))
    for idx in idxs:
        print(idx)
    count += 1
```

```python
topics = ['daytime, anxiety, depression',
          'nighttime, insomnia',
          'nerve pain, happy feelings',
          'chronic pain, cbd, creativity']
```

# t-SNE 2-D

```python
tsne = TSNE(random_state = 42)
tsne_embedding = tsne.fit_transform(nmf_embedding)
tsne_embedding = pd.DataFrame(tsne_embedding, columns = ['x', 'y'])
tsne_embedding['hue'] = nmf_embedding.argmax(axis=1)
```

```python
plt.style.use('ggplot')

fig, axs = plt.subplots(1,1, figsize=(5, 5), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .1, wspace=.001)
scatter = plt.scatter(data = tsne_embedding, x = 'x', y = 'y', s = 50, c = tsne_embedding['hue'], cmap = 'Set1')
```

# t-SNE 3-D

```python
tsne3 = TSNE(n_components = 3, random_state = 42)
tsne3_embedding = tsne3.fit_transform(nmf_embedding)
tsne3_embedding = pd.DataFrame(tsne3_embedding, columns = ['x', 'y', 'z'])
tsne3_embedding['hue'] = nmf_embedding.argmax(axis=1)
```

```python
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111, projection='3d')

xs = tsne3_embedding['x']
ys = tsne3_embedding['y']
zs = tsne3_embedding['z']
c = tsne3_embedding['hue']
ax.scatter(xs = xs, ys = ys, zs = zs, s = 100, c = c, cmap = "Set1")
```

# plotly 3-D

```python
trace = go.Scatter3d(
    x = tsne3_embedding['x'],
    y = tsne3_embedding['y'],
    z = tsne3_embedding['z'],
    mode = 'markers',
    marker = dict(
        size = 10,
        color = tsne3_embedding['hue'],                
        colorscale = 'Viridis',   
        opacity = 0.9
    )
)

plot_data = [trace]
layout = go.Layout(
    margin = dict(
        l = 0,
        r = 0,
        b = 0,
        t = 0
    )
)
fig = go.Figure(data = plot_data, layout = layout)
py.iplot(fig)
```
